```python
import pandas as pd
import helper_functions as hf
```

```python
xrp_price_df = pd.read_csv("Data/XRP.csv")
xrp_price_df = hf.multiindex('xrp',xrp_price_df)
xrp_price_df
```

```python
solana_price_df = pd.read_csv("Data/solona.csv")
solana_price_df = hf.multiindex('solana',solana_price_df)
solana_price_df
```

```python
tron_price_df = pd.read_csv("Data/tron.csv")
tron_price_df = hf.multiindex('tron',tron_price_df)
tron_price_df
```

```python
poly_price_df = pd.read_csv("Data/poly.csv")
poly_price_df = hf.multiindex('poly',poly_price_df)
poly_price_df
```

```python
fantom_price_df = pd.read_csv("Data/fantom.csv")
fantom_price_df = hf.multiindex('ftm',fantom_price_df)
fantom_price_df
```

```python
eth_price_df = pd.read_csv("Data/eth.csv")
eth_price_df = hf.multiindex('eth',eth_price_df)
eth_price_df
```

```python
cronos_price_df = pd.read_csv("Data/cron.csv")
cronos_price_df = hf.multiindex('cro',cronos_price_df)
cronos_price_df
```

```python
bnb_price_df = pd.read_csv("Data/bnb.csv")
bnb_price_df = hf.multiindex('bnb',bnb_price_df)
bnb_price_df
```

```python
total_df = xrp_price_df.merge(solana_price_df,how='left', left_index=True, right_index=True)
total_df = total_df.merge(tron_price_df,how='left', left_index=True, right_index=True)
total_df = total_df.merge(poly_price_df,how='left', left_index=True, right_index=True)
total_df = total_df.merge(fantom_price_df,how='left', left_index=True, right_index=True)
total_df = total_df.merge(eth_price_df,how='left', left_index=True, right_index=True)
total_df = total_df.merge(cronos_price_df,how='left', left_index=True, right_index=True)
total_df = total_df.merge(bnb_price_df,how='left', left_index=True, right_index=True)
total_df.dropna(inplace=True)
total_df.to_csv('all_coins2.csv')
```
